После просмотра данных встает несколько вопросов:
 * 1) Можем ли мы построить идеальную модель ? (условная метрика точность)

**Ответ**: Метрика точность в этой задачи нам подходит, тк классы сбалансированны. Но идеальную модель мы построить не сможем, тк на картинке видно, что классы перемешаны.

 * 2) Что ещё мы хотим знать о выборке?

**Ответ**: выше я узнала, что передо мной задача двухклассовой классификации с перемешанными классами, выборка сбалансированна по классам. Пропусков в данных нет.


 * 3) Какие алгоритмы машинного обучения мы можем эффективно здесь использовать?  
 
**Ответ**: для решения задачи двухклассовой классификации подходит достаточно большое количество алгоритмов, такие как байесовкий классификатор, метод решающих деревьев, бустинги, нейросети (логистическая регрессия, метод опорных веркторов,метод ближайших соседей возможно не так эффективены)

После описательного анализа приступим к построению моделей. Для начала посмотрим, что нам даст Логистическая регрессия. 

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score

# разобьем выборку на трэйн/тест
X_train, X_test, y_train, y_test = train_test_split(df[['x1','x2']], df['y'], test_size = 0.33)
```

```python
from sklearn.linear_model import LogisticRegression

#########################################
# Ваш код для Логистической регрессии
LR=LogisticRegression(random_state=123)
LR.fit(X_train,y_train)

predict_logit_train =  LR.predict(X_train)

predict_logit =  LR.predict(X_test)
predict_log_proba = LR.predict_log_proba(X_test)

#########################################

print(f'Точность нашей модели траин {accuracy_score(predict_logit_train, y_train)}')
print(f'Точность нашей модели тест {accuracy_score(predict_logit, y_test)}')
```

```python
# Заодно завизуализируем полученные результаты
plt.figure(figsize=(20, 10))
plt.scatter(X_test['x1'], X_test['x2'], c = predict_logit)
plt.show()
```

Вообще машинное обучение не только про алгоритмы (нейронные сети не исключение),
 а скорее про обработку данных. Что нам надо добавить, чтобы наш алгоритм отработал без проблем?

**Ответ:** предлагаю добавить полиномиальные признаки второго порядка.

```python
#del data
```

```python
# Повторяем наш код, только с новыми фичами - смотрим на магию
from sklearn.preprocessing import PolynomialFeatures

#########################################
# ваш код для фичей и новой модели
data=df[["x1","x2"]].copy()
poly=PolynomialFeatures(degree=2).fit(data)
data_polynom = pd.DataFrame(poly.transform(data), 
                                    columns=poly.get_feature_names_out(data.columns))
del data_polynom['1']
df_new = pd.concat([data_polynom,df['y']],axis=1)
df_new.rename(columns={'x1^2': 'new_feature_1', 'x1 x2': 'new_feature_2', 'x2^2': 'new_feature_3'}, inplace=True)
```

```python
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(df_new[['x1', 'x2', 'new_feature_1','new_feature_2','new_feature_3']],
                                                                    df_new['y'], test_size = 0.33)
LR=LogisticRegression(random_state=123)
LR.fit(X_train_new,y_train_new)

predict_logit_train =  LR.predict(X_train_new)

predict_logit =  LR.predict(X_test_new)
predict_log_proba = LR.predict_log_proba(X_test_new)

#########################################
print(f'Точность нашей модели траин {accuracy_score(predict_logit_train, y_train_new)}')
print(f'Точность нашей модели {accuracy_score(predict_logit,y_test_new)}')
```

```python
plt.figure(figsize=(20, 10))
plt.scatter(X_test_new['x1'], X_test_new['x2'], c = predict_logit)
plt.show()
```

Переходим к следующиму классу алгоритмов - деревья решений. Какая особенность есть у деревьев? Какой параметр надо зафикисировать, чтобы эта особенность не испортила обучение?

**Ответ**: для любой выборки можно построить дерево, не допускающее ни одной ошибки, но это дерево окажется переобученным. Чтобы переобучения не произошло, нужно задать критерий остонова (ограничение максимальной глубины дерева, ограничение минимального числа объектов в листьях и др. (ограничение максимального числа листьев в дереве, останов в случае, если все объекты в листе из одного класса, или требование, что функционал качества при дроблении увеличивался как минимум на сколько-то %)

```python
from sklearn.tree import DecisionTreeClassifier

#########################################
# Ваш код

dt=DecisionTreeClassifier(random_state=123)
dt.fit(X_train,y_train)

predict_tree_train = dt.predict(X_train)
print(f'Точность нашей модели train {accuracy_score(predict_tree_train, y_train)}')

predict_tree = dt.predict(X_test)
print(f'Точность нашей модели test {accuracy_score(predict_tree, y_test)}')

predict_tree_proba = dt.predict_proba(X_test)
predict_tree_proba.shape
#########################################
```

**Произошло переобучение. Посмотрим при какой ограничении максимальной глубины дерева, accuracy_score самый высокий.**

```python
max_depth_array = range(2, 20)
mse_array = []

for max_depth in max_depth_array:
    dt = DecisionTreeClassifier(max_depth=max_depth, random_state=123)
    dt.fit(X_train, y_train)
    mse_array.append(accuracy_score(y_test, dt.predict(X_test)))

plt.plot(max_depth_array, mse_array)
plt.title("Dependence of accuracy_score on max depth")
plt.xlabel("max depth")
plt.ylabel("accuracy_score")
plt.show()
```

```python
# Подгружаем tensorflow 
import tensorflow as tf
keras = tf.keras

print(tf.__version__)
print(keras.__version__) 

from tensorflow.keras.models import Sequential   # Последовательность слоёв
import tensorflow.keras.layers as L              # Разные слои
import tensorflow.keras.optimizers as opt        # Оптимизаторы
```

```python
#сделаем нормировку данных

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X_train_new)
X_train_new_sc = scaler.transform(X_train_new)
X_test_new_sc = scaler.transform(X_test_new)
```

```python
from tensorflow.keras.metrics import Accuracy

def get_model( ):
    acc = Accuracy()

    model = Sequential(name = 'MySwallow')

    model.add(L.Dense(5, input_dim = X_train_new_sc.shape[1], kernel_initializer='random_normal'))
    model.add(L.Activation('sigmoid'))
    model.add(L.Dense(5, kernel_initializer='random_normal'))
    model.add(L.Activation('sigmoid'))


    model.add(L.Dense(2, activation='softmax', kernel_initializer = 'random_normal'))
    

    optimizer = opt.Adam(learning_rate=1e-3)

    model.compile(loss = 'categorical_crossentropy', 
                  metrics=['acc'], 
                  optimizer=optimizer)
    
    return model
```

```python
model = get_model()
```

```python
model.summary()
```

```python
from tensorflow.keras.utils import to_categorical 

hist = model.fit(X_train_new_sc, to_categorical(y_train_new), validation_split=0.2, epochs=40, verbose=1)
```

```python
#########################################
# Ваш код

predict_neural = model.predict(X_test_new_sc)
predict_classes = np.round(predict_neural)

print(f'Точность нашей модели {accuracy_score(predict_classes, to_categorical(y_test_new))}')
#########################################
```

```python
plt.figure(figsize=(20, 10))
plt.scatter(X_test_new['x1'], X_test_new['x2'], c = predict_classes[:,1])
plt.show()
```
